##Load Dataset

```python
#load all raw csv into a dist
from google.colab import files
upload = files.upload()
```

```python

#import library
import pandas as pd
import io

#create dist to store dataframe
dataframes = {}

#loop to load data into dataframe
for filename, filecontent in upload.items():
    if filename.endswith(".csv"):
        df_name = filename.replace(".csv", "").lower()
        dataframes[df_name] = pd.read_csv(
            io.BytesIO(filecontent),
            encoding="latin-1",
               quotechar='"',
               sep=",", engine="python"
        )

dataframes.keys()
```

```python

sales = dataframes['sales']
products = dataframes['products']
customers = dataframes['customers']
stores = dataframes['stores']
exr = dataframes['exchange_rates']
```

```python
for name, df in dataframes.items():
    print(f"\n{name.upper()}")
    print("Shape:", df.shape)
    print("Duplicates:", df.duplicated().sum())
    print("Nulls:")
    print(df.isna().sum())
```

```python
#create function to clean column name
def clean_columns(df):
  df.columns = (df.columns.str.strip().str.lower().str.replace(' ','_'))
  return df

# customers= clean_columns(customers)
# exchange_rates = clean_columns(exchange_rates)
# products = clean_columns(products)
# sales = clean_columns(sales)
```

## Process Customer

```python
#clean column name
customers= clean_columns(customers)
```

```python
#rename cokumn name to snake cake
customers.rename(columns={'customerkey':'customer_key'},inplace=True)
customers.rename(columns={'statecode':'state_code'},inplace=True)
customers.rename(columns={'zipcode':'zip_code'},inplace=True)
```

```python
#convert birthday to datetime
customers['birthday'] = pd.to_datetime(customers['birthday'])
```

```python
#Handle missing value
#fill Null in statecode column with Unknown
customers['state_code'] = customers['state_code'].fillna('UNKNOWN')
```

```python
customers['city'] = customers['city'].str.title()
```

## Process Sales

```python
sales.info()
```

```python
#clean column name
sales = clean_columns(sales)
sales.columns
```

```python
#rename column into snake case
sales.rename(columns={'customerkey' : 'customer_key'},inplace=True)
sales.rename(columns={'storekey' : 'store_key'},inplace=True)
sales.rename(columns={'productkey' : 'product_key'},inplace=True)
```

```python
#convert datatype
sales['delivery_date'] = pd.to_datetime(sales['delivery_date'])
sales['order_date'] = pd.to_datetime(sales['order_date'])
```

```python
# checking for duplicates
duplicates = sales[sales.duplicated(subset='line_item', keep=False)]
duplicates
```

```python
sales.info()
```

```python
sales.head(10)
```

```python
sales.to_csv('cleaned_sales.csv',encoding="utf-8",index=False)
```

## Process Products

```python
products.info()
```

```python
#standardize column name
products = clean_columns(products)
```

```python
#rename column into snake case
products.rename(columns={'productkey' : 'product_key',
                         'subcategorykey' : 'subcategory_key',
                         'categorykey' : 'category_key'},inplace=True)
```

```python
#convert datatype
products['unit_price_usd'] = products['unit_price_usd'].str.replace('[$,]', '', regex=True).str.strip().astype(float)
products['unit_cost_usd'] = products['unit_cost_usd'].str.replace('[$,]', '', regex=True).str.strip().astype(float)
```

```python
# checking for duplicates
duplicates = products[products.duplicated(subset='product_key', keep=False)]
duplicates
```

```python
products['subcategory'] = products['subcategory'].str.replace(',', ' ',regex=True)
products['product_name'] = products['product_name'].str.replace(',', ' ',regex=True)
```

```python
products.info()
```

```python
products.head(10)
```

```python
products.to_csv('cleaned_products.csv',encoding="utf-8",index=False)
```
